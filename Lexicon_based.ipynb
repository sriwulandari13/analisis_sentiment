```python
import pandas as pd
import io

df = pd.read_csv('preprocessing.csv', encoding='windows-1252')
```

```python
df.head(5)
```

```python
#lexicon based
!pip install VaderSentiment
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyser = SentimentIntensityAnalyzer()
```

```python
scores = [analyser.polarity_scores(x) for x in df['Text']]
print(scores)
df['Compound_Score'] = [x['compound'] for x in scores]
```

```python
df.head(100)
```

```python
#compound_score
df.loc[df['Compound_Score'] < 0, 'Sentiments'] = 'Negatif'
df.loc[df['Compound_Score'] == 0, 'Sentiments'] = 'Netral'
df.loc[df['Compound_Score'] > 0, 'Sentiments'] = 'Positif'
df.head(100)
```

```python
#save
df.to_csv('hasil_lexiconbased', encoding='utf-8', index=False)
```
